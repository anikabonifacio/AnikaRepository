```python
import requests
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
i = 0
latitudes =[]
longitudes = []
while i < 100:
    data = requests.get('https://api.wheretheiss.at/v1/satellites/25544').json()
    print(i+1)
    print("Latitude = ", data['latitude'])
    print("Longitude = ", data['longitude'])
    print("Timestamp = ", data['timestamp'])
    latitudes.append(data['latitude'])
    longitudes.append(data['longitude'])
    i += 1
    time.sleep(10)
plt.figure(figsize=(10, 5))
plt.scatter(longitudes, latitudes, color='green', marker='o')

plt.title('ISS Path Tracker')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.grid(True)
plt.show()
```
